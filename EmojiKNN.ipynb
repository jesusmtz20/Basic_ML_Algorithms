#Algoritmo K-Nearest Neighbors para emojis asociados a frases

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report
```

```python
data = pd.read_csv('frases_emojis.csv')
data.info()
```

##Limpieza de datos


```python
def preprocess_text(frase):
    return frase.lower()
    #return frase.lower()
data['frases_procesadas'] = data['frase'].apply(preprocess_text)
```

##VectorizaciÃ³n de frases con TF-IDF

```python
tfidf = TfidfVectorizer()
X = tfidf.fit_transform(data['frases_procesadas'])
Y = data['emoji']
```

##Entrenamiento del modelo

```python
#Datos para entrenamiento y pruebas
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=10)

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, Y_train)
```

##Predicciones en el conjunto de pruebas

```python
Y_pred = knn.predict(X_test)
```

```python
test_data = pd.DataFrame({'Frase': data['frase'][Y_test.index],
                          'Emoji real': Y_test,
                          'Emoji previsto': Y_pred})

print("Emojis correctos:")
print(test_data[test_data['Emoji real'] == test_data['Emoji previsto']])

print("Emojis incorrectos:")
print(test_data[test_data['Emoji real'] != test_data['Emoji previsto']])
```

##AnÃ¡lisis de resultados

La precisiÃ³n del modelo para el conjunto de datos de prueba es de 38%, lo cual es razonable considerando la cantidad y poca representaciÃ³n con frases de algunos emojis utilizados.

Para mejorar la precisiÃ³n, es necesario aumentar el tamaÃ±o y diversidad del conjunto de datos con mÃ¡s frases representativas de cada emoji. Esto se refleja en el nÃºmero de ejemplos que utilizÃ³ el modelo para cada emoji (mÃ©trica support).

Finalmente, es importante considerar que los resultados obtenidos estan influenciados por diferentes factores como el mÃ©todo para vectorizar las frases, el modelo de machine learning utilizado y el hecho de que con un solo emoji se pueden expresar diferentes ideas o emociones. Para mejorar los resultados obtenidos, se deben analizar y ajustar estos factores.

```python
print(classification_report(Y_test, Y_pred, zero_division=0))
```
